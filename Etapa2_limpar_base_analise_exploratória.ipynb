```python
# vamos modificar o título da variável com parênteses, pois parênteses no nome costuma dar problema
```

```python
df2=df2.rename(columns={'Valor_do_aluguel_(em_reais)':'Valor_do_aluguel_em_reais'})
```

```python
df2=df2.rename(columns={'Nº_de_comodos':'N_de_comodos'})
```

```python
# abaixo vemos os resultados da regressão e aquelas variáveis que podemos considerar minimamente relevantes
```

```python
result = sm.ols(formula="Valor_do_aluguel_em_reais~N_de_comodos+Comodos_como_dormitorio_numero+Regiao_0+Regiao_1+Regiao_2+Regiao_3+Regiao_4+Regiao_5+Regiao_6+Regiao_7+Regiao_8", data=df2).fit()
print(result.summary())
```
